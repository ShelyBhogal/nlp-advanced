# Basics of TensorFlow

In this demonstration of TensorFlow, you will create a synthetic dataset of randomly-selected values, specifically for a Linear Regression task. This is to show the basic syntax of TensorFlow as it builds the 'line of best fit'.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

```python
# Create input X and output y variables that have linear relationship

# No of samples
N = 100

# Uniform random distribution of 100 values between -3 and 3
X = np.random.random(N) * 6 - 3

# Linear function of X plus some Gaussian-distributed 'noise'
y = 0.5 * X - 1 + np.random.randn(N) * 0.5
```

**NOTE: The slope (0.5) and y-intercept (-1) of the line in the y variable should be remembered, since one of your goals is to recover these values in the model results.**

```python
plt.scatter(X, y);
```

**You can see the linear shape of the plot, as well as some noise**

## Build the model

`Input()` function defines the inner layer of a neural network, i.e. the input data - is it one-dimensional, two-dimensional etc.

`Dense()` function defines the hidden layer of a neural network, with the number of neurons.

`Model()` function defines the deep learning model by the inner and hidden layers.

```python
# Input layer data is one-dimensional, i.e. x variable
i = Input(shape=(1,))

# Implement dense layer with one neuron attached to inner layer
x = Dense(1)(i)

# Model is made of input and densely-connected layer
model = Model(i, x)
```

```python
model.summary()
```

When compiling a linear regression model, you need to define the parameters on how the model is trained before introducing any data. The most important parameters are:

* `optimizer` - specify the type of Gradient Descent. Linear Regression tasks accept Adam, 
* `loss` - specify function to measure losses, like MSE, RMSE, MAE etc.
* `metrics` - additional performance metrics like MSE, RMSE, MAE etc.

When compiling a linear regression model, it is usually pointless including a metric since we only care about the test MSE score.

```python
# You can use Adam() function to set learning rate (instead of using optimizer='adam')

model.compile(loss='mse', optimizer=Adam(learning_rate=0.1), metrics=['mae'])
```

```python
r = model.fit(X.reshape(-1, 1), y, epochs=200, batch_size=32)
```

```python
# History object (dictionary-like) records training loss & metric (MAE) at each epoch

r.history
```

## Assess model performance

```python
# Plot training losses per epoch

plt.plot(r.history['loss'], label='loss')
plt.legend();
```

```python
# This is what we want to see - after 12 epochs the training loss levels out around 0.25
```

```python
# Plot training MAE per epoch

plt.plot(r.history['mae'], label='MAE')
plt.legend();
```

```python
# Similar to MSE, the MAE levels out after approx 13 epochs which is what we want to see
```

```python
# Make predictions (create test set 20 evenly-spaced values between -3 and 3)

X_test = np.linspace(-3, 3, 20).reshape(-1, 1)

predictions = model.predict(X_test)
```

```python
# Original dataset
plt.scatter(X, y)

# Line of best fit
plt.plot(X_test, predictions);
```

**The 'line of best fit' passes nicely through the training data points.**

## Check the 'learned' parameters

```python
# List of layers in model (can index)

model.layers
```

```python
# Slope is approx 0.59 and intercept is approx -1.1

model.layers[1].get_weights()
```

**The slope and intercept values, i.e. weights of hidden layer, closely match the ones in the original *y* target variable, so we know the model is working as expected.**
